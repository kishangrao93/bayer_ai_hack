```python
import boto3, json

# Initialize Bedrock client
client = boto3.client("bedrock-agent-runtime")

def retrieve_kb(query_text: str, knowledge_base_id: str, max_results: int = 10):
    """
    Minimal working KB retrieval from AWS Bedrock.
    """
    response = client.retrieve(
        knowledgeBaseId=knowledge_base_id,
        retrievalQuery={
                "text": query_text,
                "type": "TEXT"
            },
        retrievalConfiguration={
            "vectorSearchConfiguration": {
                "numberOfResults": max_results
            }
        }
    )
    all_queries = []
    kb_snippets = response['retrievalResults']
    for snippet in kb_snippets:
        all_queries.append(snippet['content']['text'])
    return all_queries
```
